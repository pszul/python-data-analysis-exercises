```python
import numpy as np
```

```python
## make 4x4 matrix
A = np.random.uniform(size=16, low=-10, high=10).reshape(4,4)

A
```

```python
## slice 2x2 from inside
A[1:3,1:3]
```

```python
## values above first, second, third quantile
def show_values_above_quantile(x, I):
    q = np.quantile(I, x)
    mask_q1 = I>q
    return q, np.sort(I[mask_q1])

display(f"first quantile = {show_values_above_quantile(0.25,A)[0]:.2f}")
display(np.round(show_values_above_quantile(0.25,A)[1],2))

display(f"second quantile = {show_values_above_quantile(0.5,A)[0]:.2f}")
display(np.round(show_values_above_quantile(0.5,A)[1],2))

display(f"third quantile = {show_values_above_quantile(0.75,A)[0]:.2f}")
display(np.round(show_values_above_quantile(0.75,A)[1],2))
```

```python
# mask all values below zero as 0
mask = A<0
A[mask] = 0

display(np.round(A,2))
```

```python
# swapping columns in matrix
A_swapped =  A[:,[3,1,2,0]]

display(np.round(A_swapped,2))
```
