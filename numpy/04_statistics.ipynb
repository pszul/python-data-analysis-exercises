```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## simulate 1000 coin tosses
probability = 0.5
n_of_tosses = 1000
number_of_samples = 100

bernoulli_has_more_heads = np.array([], dtype=bool)


for _ in range(number_of_samples):
    x_choice = np.random.choice([0,1], size=n_of_tosses, p=[1-probability, probability])
    x_bernoulli = np.random.binomial(1, p=probability, size=n_of_tosses)
    no_of_heads_choice = x_choice[x_choice == 1].sum()
    no_of_heads_bernoulli = x_bernoulli[x_bernoulli == 1].sum()
    bernoulli_has_more_heads = np.append(bernoulli_has_more_heads,
                                         no_of_heads_bernoulli < no_of_heads_choice)

count_true = bernoulli_has_more_heads.sum()

if count_true > 50:
    print(f"Bernoulli has more heads: {count_true}")
elif count_true == 50:
    print(f"Meh, the same")
else:
    print(f"\"Choice\", has more heads: {number_of_samples - count_true}")
```

```python
# gaussian distribution

s=1000

gaussian_distribution = np.random.normal(loc=0.0,
                                         scale=1,
                                         size=s)

mu = gaussian_distribution.mean()
sigma = gaussian_distribution.std(ddof=1)


x = np.linspace(gaussian_distribution.min(), gaussian_distribution.max(), 400)
pdf = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-0.5*((x-mu)/sigma)**2)

plt.figure(figsize=(8, 4.5))
plt.hist(gaussian_distribution, bins=15, density=True, alpha=0.6, edgecolor='black')
plt.plot(x, pdf, linewidth=2)

plt.title(f"Gaussian distribution sample (n={s})")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True, alpha=0.3)
plt.show()

```
