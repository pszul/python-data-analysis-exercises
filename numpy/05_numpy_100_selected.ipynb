22. Normalize a 5x5 random matrix (★☆☆)

```python
matrix = np.random.uniform(size=25).reshape(5,5)

mean_cols = matrix.mean(axis=0)
std_cols = matrix.std(axis=0,ddof=0)

matrix_norm = (matrix - mean_cols) / std_cols

display(np.round(matrix_norm.mean(axis=0)))
display(np.round(matrix_norm.std(axis=0)))
display(matrix_norm)
```

41. How to sum a small array faster than np.sum? (★★☆)

```python
big_array = np.random.uniform(low=0.0, high=1.0,size=10000)
small_array = np.random.uniform(low=0.0, high=1.0,size=10)


%time big_array.sum()
%time np.sum(big_array)
%time np.add.reduce(big_array)

%time small_array.sum()
%time np.sum(small_array)
%time np.add.reduce(small_array)
```

94. Considering a 10x3 matrix, extract rows with unequal values (e.g. [2,2,3]) (★★★)

```python
A = np.random.uniform(size=30)*10

A = np.round(A).reshape(10,3)
A = np.astype(A,int)

# Adding equal rows to test (up to 7)
how_much_equal_rows = np.random.randint(1, 8)

# rows to replace
rows_to_replace = np.random.choice(np.arange(1, 10), size=how_much_equal_rows, replace=False)

for i in rows_to_replace:
    A[i] = [i,i,i]

display(A)

unequal_rows = A[np.ptp(A, axis=1) != 0]

display("Unequal rows" ,unequal_rows)
```
