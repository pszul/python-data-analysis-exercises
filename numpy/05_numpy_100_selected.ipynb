```python
matrix = np.random.uniform(size=25).reshape(5,5)

mean_cols = matrix.mean(axis=0)
std_cols = matrix.std(axis=0,ddof=0)

matrix_norm = (matrix - mean_cols) / std_cols

display(np.round(matrix_norm.mean(axis=0)))
display(np.round(matrix_norm.std(axis=0)))
display(matrix_norm)
```

51. Create a structured array representing a position (x,y) and a color (r,g,b) (★★☆)

```python
position_and_color = np.dtype([
    ('position' , float , 2),
    ('RGB'      , int   , 3)
])

no_of_points = np.random.randint(10, 100)
x = np.random.uniform(low=0.0, high=100.0, size=no_of_points)
y = np.random.uniform(low=0.0, high=100.0, size=no_of_points)
R = np.random.randint(0, 256, size=no_of_points)
G = np.random.randint(0, 256, size=no_of_points)
B = np.random.randint(0, 256, size=no_of_points)


color_points = np.zeros(no_of_points, dtype=position_and_color)

color_points['position'] = np.column_stack((x, y))
color_points['RGB']      = np.column_stack((R, G, B))
colors = color_points['RGB'] / 255.0

plt.figure(figsize=(6, 6))
plt.scatter(
    color_points['position'][:, 0],
    color_points['position'][:, 1],
    c=colors,
    s=100
)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Random colored points")
plt.grid(True)
plt.tight_layout()
plt.show()

color_points
```

63. Create an array class that has a name attribute (★★☆)

```python
class NamedArray(np.ndarray):
    def __new__(cls, input_array, name=""):
        obj = np.asarray(input_array).view(cls)
        obj.name = name
        return obj

    def __array_finalize__(self, obj):
        if obj is None:
            return
        self.name = getattr(obj, 'name', None)

array = np.random.uniform(low=0.0, high=1.0,size=10000)

a = NamedArray(array, name="my_vector")

a, a.name
```

70. Consider the vector [1, 2, 3, 4, 5], how to build a new vector with 3 consecutive zeros interleaved between each value? (★★★)

```python
vector = np.array([1, 2, 3, 4, 5])
k = 3

no_of_elements = len(vector)

out = np.zeros(no_of_elements + (no_of_elements - 1) * k, dtype=int)

out[::k+1] = vector

vector, out
```

83. How to find the most frequent value in an array?

```python
a = np.random.randint(2,10)
b = np.random.randint(2,10)
c = np.random.randint(10,100)

if a > b:
    a = a + b
    b = a - b
    a = a - b

A = np.random.randint(a,b,c)

val , counts = np.unique(A, return_counts=True)
most_frequent_val = val[counts.argmax()]

A, most_frequent_val, counts.max()
```

94. Considering a 10x3 matrix, extract rows with unequal values (e.g. [2,2,3]) (★★★)

```python
A = np.random.uniform(size=30)*10

A = np.round(A).reshape(10,3)
A = np.astype(A,int)

# Adding equal rows to test (up to 7)
how_much_equal_rows = np.random.randint(1, 8)

# rows to replace
rows_to_replace = np.random.choice(np.arange(1, 10), size=how_much_equal_rows, replace=False)

for i in rows_to_replace:
    A[i] = [i,i,i]

display(A)

unequal_rows = A[np.ptp(A, axis=1) != 0]

display("Unequal rows" ,unequal_rows)
```
