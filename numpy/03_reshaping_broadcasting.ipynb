```python
import numpy as np
```

```python
## reshape
I = np.random.uniform(size=12, low=-10, high=10)
A = I.reshape(3,4)
B = I.reshape(2,6)

np.round(A,2),np.round(B,2)
```

```python
## broadcasting
V = np.random.uniform(size=3, low=-10, high=10)

A_plus_v = A + V[:, None]

display(V)
display(A)
display(A_plus_v)
```

```python
# normalisation
mean_cols = A.mean(axis=0)
std_cols = A.std(axis=0,ddof=0)

A_norm = (A - mean_cols) / std_cols

display(A_norm)
```
