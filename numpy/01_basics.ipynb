```python
import numpy as np
```

```python
# Generation examples
arr = np.arange(9)
mtx = np.random.uniform(size=9).reshape(3,3)
I = np.eye(4)

arr, mtx, I
```

```python
random_vector = np.random.uniform(size=100)
random_vector = random_vector * np.random.uniform(size=100, low=0, high=100)
```

```python
# Describe

display(f"sum = {random_vector.sum():.2f}")
display(f"mean = {random_vector.mean():.2f}")
display(f"median = {np.median(random_vector):.2f}")
display(f"standard_deviation = {random_vector.std():.2f}")

```
